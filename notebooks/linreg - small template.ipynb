```python
from scipy.misc import lena
from scipy import ndimage
```

```python
image = ndimage.zoom(lena(), 0.50)
r, c = image.shape
template = ndimage.rotate(image, 20, reshape=False)[(r/2)-20:(r/2)+20, (c/2)-20:(c/2)+20]
```

```python
fix, ax1 = plt.subplots(1, 1, figsize=(8, 8))
ax1.imshow(image, cmap='gray')
_ = ax1.set_title('image')

fix, ax2 = plt.subplots(1, 1, figsize=(2, 2))
ax2.imshow(template, cmap='gray')
_ = ax2.set_title('template')
```

```python
from imreg import register
```

```python
template = register.RegisterData(template)

image = register.RegisterData(image)
```

Process
-------

Define a template "rectangle", which is transformed in image space.

Define a resampling function for the rectangles.


```python
def drawRect(ax, rectangle, style='--r'):
    ax.plot( rectangle[0], rectangle[1], style )
    ax.plot( rectangle[1], rectangle[3], style )
    ax.plot( rectangle[3], rectangle[2], style )
    ax.plot( rectangle[2], rectangle[0], style )
    
```

```python
fix, ax1 = plt.subplots(1, 1, figsize=(8, 8))
ax1.imshow(image.data, cmap='gray')
ax1.axis('tight')
drawRect(ax1, template.coords.bbox)
_ = ax1.set_title('image')
```

```python
from imreg import sampler
```

```python
sampler.bilinear(image.data, template.coords.tensor)
```

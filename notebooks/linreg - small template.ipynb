```python
from imreg import register, sampler, model, metric
```

```python
from scipy.misc import lena
from scipy import ndimage
```

```python
image = ndimage.zoom(lena(), 0.1)
template = ndimage.rotate(image, 30, reshape=False)
```

```python
registrator = register.Register(model.Affine, metric.Residual, sampler.Bilinear)
```

```python
image = register.RegisterData(image)
template = register.RegisterData(template)
```

```python
step, warpedImage, search = registrator.register(
    image,
    template,
    verbose=True,
    )
```

```python
%timeit registrator.register(image, template, verbose=False)
```

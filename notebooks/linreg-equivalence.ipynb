```python
from imreg import register, sampler, model, metric
```

```python
from scipy.misc import lena
from scipy import ndimage
```

```python
image = ndimage.zoom(lena(), 0.1)
template = ndimage.rotate(image, 30, reshape=False)
```

```python
registrator = register.Register()
```

```python
image = register.RegisterData(image)

template = register.RegisterData(template)

tform = model.Affine()
```

```python
step, additive_search = registrator.register(image, template, tform, sampler=sampler.bilinear, method=metric.forwardsAdditive)

step, compositional_search = registrator.register(image, template, tform, sampler=sampler.bilinear, method=metric.forwardsCompositional)

step, inverse_compositional_search = registrator.register(image, template, tform, sampler=sampler.bilinear, method=metric.inverseCompositional)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 5))

ax.plot( 
    [x.error for x in additive_search if x.decreasing], 's-y', 
    [x.error for x in compositional_search if x.decreasing], '^-r',
    [x.error for x in inverse_compositional_search if x.decreasing], '^-g',
    )

ax.grid()
_ = ax.legend(['forwards-additive', 'forwards-compositional', 'inverse-compositional'])
    
```
